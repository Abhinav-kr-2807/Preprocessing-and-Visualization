##**3.drop all the non-null value columns from the dataframe**

```python
dz1=dz.drop(columns=['Loan_ID', 'Dependents', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History'])
dz1
```

```python
dz1.columns
```

##**4.fill the null values with categorical imputer**

```python
# using sklearn-pandas package
from sklearn_pandas import CategoricalImputer
 
# handling NaN values
imputer = CategoricalImputer()


Gender = np.array(dz1['Gender'], dtype=object)
Gender_fill=imputer.fit_transform(Gender)


Married = np.array(dz['Married'], dtype=object)
Married_fill=imputer.fit_transform(Married)

Self_Employed = np.array(dz['Self_Employed'], dtype=object)
Self_Employed_fill=imputer.fit_transform(Self_Employed)


print(Gender_fill)
print(Married_fill)
print(Self_Employed_fill)
```

```python
#dz2=pd.DataFrame(data='Gender_fill, 'Married_fill, Self_Employed_fill)
#dz2
Dictionary_2 = {'Gender_fill' : Gender_fill,
  'Married_fill' : Married_fill,
  'Self_Employed_fill' : Self_Employed_fill}
dz2=pd.DataFrame(Dictionary_2) 
dz2
```

##**5.create new dataframe with all previously non-null columns and newly fill value columns**

```python
dz3=pd.concat([dz1,dz2],axis=1)

dz3
```

```python
dz3.columns
```

```python
dz3.drop(columns=['Gender', 'Married', 'Self_Employed'],inplace=True)
dz3
```

```python
dz3.isna().sum()
```

```python
dz3.columns
```

##**6.use the lable encoder to encode the categorical data**

```python
dz3['Education_labels']= label_encoder.fit_transform(dz3['Education']) 
dz3['Property_Area_labels']= label_encoder.fit_transform(dz3['Property_Area'])
dz3['Loan_Status_labels']= label_encoder.fit_transform(dz3['Loan_Status'])
dz3['Gender_labels']= label_encoder.fit_transform(dz3['Gender_fill']) 
dz3['Married_labels']= label_encoder.fit_transform(dz3['Married_fill']) 
dz3['Self_Employed_labels']= label_encoder.fit_transform(dz3['Self_Employed_fill']) 
dz3
```

```python
dz3.drop(columns=['Education', 'Property_Area', 'Loan_Status', 'Gender_fill',
       'Married_fill', 'Self_Employed_fill'], inplace=True)
dz3
```

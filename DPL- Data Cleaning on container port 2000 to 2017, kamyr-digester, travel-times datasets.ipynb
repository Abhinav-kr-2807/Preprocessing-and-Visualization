```python
df2=pd.read_csv("kamyr-digester.csv")
df2
```

```python
df2.isnull()
```

```python
df2.isnull().sum().sum()
```

```python
df2.dtypes
```

```python
df2 = df2.select_dtypes(exclude=['object'])
df2
#removing columns with object dtype
```

```python
before_imputation=pd.DataFrame(df2)
before_imputation
```

```python
imputer=KNNImputer(n_neighbors=2)
After_imputation=imputer.fit_transform(before_imputation)
After_imputation
```

```python
df2=pd.DataFrame(After_imputation)
df2
#KNNImputer removes the column names
```

```python
before_imputation.columns
```

```python
After_imputation=pd.DataFrame(After_imputation,columns=before_imputation.columns)
After_imputation
#adding columns in final dataframe
```

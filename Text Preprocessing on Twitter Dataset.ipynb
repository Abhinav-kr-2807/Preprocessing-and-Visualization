```python
import pandas as pd
import numpy as np
import nltk

from nltk import *
nltk.download()
```

```python
df = pd.read_csv('twitter dataset.csv')
df.head(20)
```

```python
df.shape
```

## Text Cleaning

```python
# lowering alphabets
df['tweet'] = pd.Series([w.lower() for w in df['tweet']])
df.head(20)
```

```python
# removing hash tags
df['tweet'] = df['tweet'].apply(lambda x: (re.sub("(#\w+)", "", x)))
df.head(20)
```

```python
# removing html tags

pattern = r"(<\w+>\s?\w+>)|(</\w+>)"
df['tweet'] = df['tweet'].apply(lambda x: (re.sub(pattern, "", x)))
df.head(20)
```

```python
# removing tags @

pattern1 = r"(@\s?\w+\b)"
df['tweet'] = df['tweet'].apply(lambda x: (re.sub(pattern1, "", x)))
df.head(20)
```

```python
# removing punctuations and stopwords
import re

df['tweet'] = pd.Series([re.sub(r"[^\w\s]", "", w) for w in df['tweet']])

df.head(10)
```

```python
# removing digits

df['tweet'] = df['tweet'].apply(lambda x: (''.join([a for a in x if a.isalpha() or a.isspace()])))
df.head(20)
```

## Tokenization

```python
df['tokenized_tweet'] = df['tweet'].apply(lambda x: (nltk.word_tokenize(x)))
df.head(20)
```

## Normalization

```python
#Lemmatization

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


df['tokenized_tweet'] = df['tokenized_tweet'].apply(lambda x: [lemmatizer.lemmatize(w) for w in x])
```

```python
df.head(20)
```

```python
df['tweet'] = df['tokenized_tweet'].apply(lambda x: " ".join(a for a in x))
df.head(20)
```

```python
df = df.drop('tokenized_tweet', axis = 1)
```

```python
df.head(20)
```

## Vectorization and Machine Learning Model
#### Tfidf Vectorizer
#### Multinomial Naive Bayes' Model
#### ROC_AUC_Score

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(df['tweet'], 
                                                    df['label'], 
                                                    random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import roc_auc_score
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df = 3)
vect = vectorizer.fit(X_train)
X_train_transformed = vect.transform(X_train)

X_test_transformed = vectorizer.transform(X_test)

clf = MultinomialNB(alpha=0.1)
clf.fit(X_train_transformed, y_train)

y_predicted = clf.predict(X_test_transformed)
    
print(roc_auc_score(y_test, y_predicted))
```
